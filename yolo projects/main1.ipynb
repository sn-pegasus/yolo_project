```python
cap =  cv2.VideoCapture(vedio_path)
box_annotator =  sv.BoxAnnotator(color_palette)
print(box_annotator)
frame_count = 0
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break  # Exit loop if video ends or there's an issue

    frame_count += 1  # Update frame index

    # Run YOLOv8 inference on the frame
    results = model(frame)
    type(results)
    # Extract detections
    for result in results:
        # detections = sv.Detections.from_ultralytics(result)
        boxes = result.boxes.xyxy.cpu().numpy()  # Get bounding boxes
        confidences = result.boxes.conf.cpu().numpy()  # Get confidence scores
        class_ids = result.boxes.cls.cpu().numpy().astype(int)  # Get class IDs

        # Manually create `sv.Detections` (skip `id` field)
        detections = sv.Detections(
            xyxy=boxes,
            confidence=confidences,
            class_id=class_ids
        )
        # Assign labels with class names and confidence scores
        # labels = [f"{model.model.names[class_id]} {conf:.2f}" for class_id, conf in zip(result.boxes.cls, result.boxes.conf)]
        # labels = [f"{model.model.names[int(class_id)]} {float(conf):.2f}" for class_id, conf in zip(result.boxes.cls.cpu().numpy(), result.boxes.conf.cpu().numpy())]

        # Annotate the frame with detections
        # frame = box_annotator.annotate(scene=frame, detections=detections, labels=labels)
        labels4 = [
            f"{model.model.names[class_id]} {conf:.2f}"
            for class_id, conf in zip(class_ids, confidences)
        ]

        # Annotate the frame
        frame = box_annotator.annotate(scene=frame, detections=detections, labels=labels4)
    # Display the processed frame
    cv2.imshow("CCTV Object Detection", frame)

    # Press 'q' to exit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release video capture and close windows
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import ultralytics
from ultralytics import YOLO
import supervision as sv
import numpy as np
import os
```

```python
# Open video capture (live feed or video file)
cap = cv2.VideoCapture(0)  # Change to a video file path for pre-recorded video

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Perform object detection
    results = model(frame)

    # Render the results on the frame
    frame = results.render()[0]

    # Display the frame
    cv2.imshow("Live Object Detection", frame)

    # Press 'q' to exit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release resources
cap.release()
cv2.destroyAllWindows()
```
